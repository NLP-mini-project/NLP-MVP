```python
def titles():
    inshorts = []

    for i in range(1,10):

        url = f'https://github.com/search?p={i}&q=covid-19&type=Repositories'
        response = get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        tails = soup.find_all('a', class_ = 'v-align-middle')
        titles = [title.text for title in tails]

        inshorts.append(titles)
        
    return inshorts
```

```python
titles()
```

```python
REPOS = []

for i in range(1,15):

    url = f'https://github.com/search?p={i}&q=covid-19&type=Repositories'
    response = get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    tails = soup.find_all('a', class_ = 'v-align-middle')
    titles = [title.text for title in tails]
    
    time.sleep(3)

    REPOS.extend(titles)
```
